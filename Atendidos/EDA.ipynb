```python
unicos  = df12.drop_duplicates(subset=['COD_USU'])
```

## ACTIVO O INACTIVO

```python
conteo_discapacidad = unicos['EST_ACT'].replace({1: 'Activo', 2: 'Inactivo'}).value_counts().sort_index()

ax = sns.barplot(x=conteo_discapacidad.index, y=conteo_discapacidad.values, color='skyblue')
for container in ax.containers:
    ax.bar_label(container)

plt.title('Cantidad de Activos')
plt.xlabel('¿Activo?')
plt.ylabel('Cantidad de Personas')

plt.show()
```

Podemos ver que más de la mitad de los usarios son activos, por lo que se hará un análisis general y a solo los activos

```python
activos = unicos.query('EST_ACT == 1')
```

```python
sexo_labels = unicos['SEX_USU'].replace({1: 'Hombre', 2: 'Mujer'})
sexo_labels_a = activos['SEX_USU'].replace({1: 'Hombre', 2: 'Mujer'})

fig, axes = plt.subplots(1, 2, figsize=(14, 5))


sns.countplot(x=sexo_labels, hue=sexo_labels, ax=axes[0])

for container in axes[0].containers:
    axes[0].bar_label(container)

axes[0].set_title('Distribución por Sexo de todos los registrados')
axes[0].set_xlabel('Sexo')
axes[0].set_ylabel('Cantidad de Personas')

sns.countplot(x=sexo_labels_a, hue=sexo_labels_a, ax=axes[1])
for container in axes[1].containers:
    axes[1].bar_label(container)

axes[1].set_title('Distribución por Sexo de los activos')
axes[1].set_xlabel('Sexo')
axes[1].set_ylabel('Cantidad de Personas')

plt.tight_layout()
plt.show()
```

Podemos evidenciar que la cantidad de hombres son 2176 y la canitdad de mujeres son 2190, lo cual evidencia que el servicio de cuidado diurno atendía por igual a niños y niñas. Dentro de los usuarios activos, podemos ver que la distribución sigue siendo la misma, pero hay más mujeres que hombres.

## Étnica

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))


diccionario_lenguas = {
    1: 'Quechua', 2: 'Aimara', 3: 'Asháninka', 4: 'Awajún/Aguaruna', 
    5: 'Shipibo-Conibo', 6: 'Shawi/ Chayahuita', 7: 'Matsigenka/ Machiguenga', 
    8: 'Achuar', 9: 'Otra lengua indígena u originaria', 10: 'Castellano', 
    11: 'Portugués', 12: 'Otra lengua extranjera', 13: 'Lengua de seña peruana', 14: 'Es sordomudo / no habla', 15: 'SIN DATO'
}

conteo_lengua = unicos['LEN_MAT'].replace(diccionario_lenguas).value_counts()

ax = sns.barplot(x=conteo_lengua.index, y=conteo_lengua.values, color='skyblue', ax=axes[0])
for container in ax.containers:
    ax.bar_label(container)

axes[0].set_title('Cantidad de Usuarios Registrados por Lengua Materna')
axes[0].set_xlabel('Lengua Materna')
axes[0].set_ylabel('Cantidad de Personas')
axes[0].tick_params(axis='x', rotation=45) 

conteo_lengua_a = activos['LEN_MAT'].replace(diccionario_lenguas).value_counts()

ax_a = sns.barplot(x=conteo_lengua_a.index, y=conteo_lengua_a.values, color='skyblue', ax=axes[1])
for container in ax_a.containers:
    ax_a.bar_label(container)

axes[1].set_title('Cantidad de Usuarios Activos por Lengua Materna')
axes[1].set_xlabel('Lengua Materna')
axes[1].set_ylabel('Cantidad de Personas')
axes[1].tick_params(axis='x', rotation=45) #

plt.tight_layout()
plt.show()
```

Se puede ver que la mayoría de los usuarios tienen como lengua materna el Castellano, un insight interesante es que en los registros totales y en los registros activos, la cantidad de usuarios de la categoría 'Sin Dato' no cambia. Se puede intuir que son nuevos ingresantes y que a otros ingresantes se les asigna una lengua materna luego de estar cierto tiempo con ellos.

```python
unicos['FEC_ING'] = pd.to_datetime(unicos['FEC_ING'])
unicos['MES_AÑO_ING'] = unicos['FEC_ING'].dt.strftime('%Y-%m')

conteo_temporal = unicos.groupby(['MES_AÑO_ING', 'LEN_MAT']).size().reset_index(name='Cantidad')
conteo_temporal['LEN_MAT'] = conteo_temporal['LEN_MAT'].replace(diccionario_lenguas)

tabla_pivot = conteo_temporal.pivot(index='MES_AÑO_ING', columns='LEN_MAT', values='Cantidad').fillna(0)

ax = tabla_pivot.plot(kind='line', marker='o', figsize=(16, 8))

plt.title('Evolución de Ingresos (Mes y Año) por Lengua Materna')
plt.xlabel('Mes y Año de Ingreso')
plt.ylabel('Cantidad de Ingresantes')

plt.xticks(range(len(tabla_pivot.index)), tabla_pivot.index, rotation=90)
plt.legend(title='Lengua Materna', bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Se puede ver que el ingreso de las personas de 'Sin Dato' son recientes, eso explicaría por qué se han quedado Sin Dato en  la Lengua Materna. Además podemos ver que en ingreso de personas que hablan Aimara como lengua materna se incremento radicalmente en el mes de enero y marzo de 2025. 

Por otro lado, podríamos tener como hipótesis que como son niños de edades muy tempranas, no pueden determinar la lengua materna.

```python
diccionario_edades = {
    1: '0 - 11 meses', 2: '1 - 5 años', 3: '6 - 11 años', 4: '12 - 17 años', 5: '18 años'
}

datos_grafico = unicos[['GRU_ET', 'LEN_MAT']].copy()
datos_grafico['GRU_ET'] = datos_grafico['GRU_ET'].replace(diccionario_edades)
datos_grafico['LEN_MAT'] = datos_grafico['LEN_MAT'].replace(diccionario_lenguas)

orden_edades = ['0 - 11 meses', '1 - 5 años', '6 - 11 años', '12 - 17 años', '18 años']

plt.figure(figsize=(14, 7))

ax = sns.countplot(data=datos_grafico, x='GRU_ET', hue='LEN_MAT', order=orden_edades, palette='tab20')

for container in ax.containers:
    ax.bar_label(container)

plt.title('Distribución de Lengua Materna por Grupo Etario')
plt.xlabel('Grupo Etario')
plt.ylabel('Cantidad de Personas')

plt.legend(title='Lengua Materna', bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

## Nombre de los centros

```python
conteo_centros = unicos.groupby('NOM_CEN').size().reset_index(name='Cantidad').sort_values('Cantidad', ascending=False)

conteo_centros_a = activos.groupby('NOM_CEN').size().reset_index(name='Cantidad').sort_values('Cantidad', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(18, 12))

ax1 = sns.barplot(data=conteo_centros, x='Cantidad', y='NOM_CEN', color='skyblue', ax=axes[0])
for container in ax1.containers:
    ax1.bar_label(container, padding=3)

axes[0].set_title('Usuarios Registrados por Centro')
axes[0].set_xlabel('Cantidad de Personas')
axes[0].set_ylabel('Nombre del Centro')

# Gráfico 2: Usuarios Activos (Derecha)
ax2 = sns.barplot(data=conteo_centros_a, x='Cantidad', y='NOM_CEN', color='skyblue', ax=axes[1])
for container in ax2.containers:
    ax2.bar_label(container, padding=3)

axes[1].set_title('Usuarios Activos por Centro')
axes[1].set_xlabel('Cantidad de Personas')
axes[1].set_ylabel('') 

plt.tight_layout()
plt.show()
```

```python
total_por_centro = unicos.groupby('NOM_CEN').size().reset_index(name='Total_Registrados')
activos_por_centro = activos.groupby('NOM_CEN').size().reset_index(name='Total_Activos')
retencion = pd.merge(total_por_centro, activos_por_centro, on='NOM_CEN', how='left')
retencion['Total_Activos'] = retencion['Total_Activos'].fillna(0)

retencion['Porcentaje_Activos'] = (retencion['Total_Activos'] / retencion['Total_Registrados']) * 100

retencion = retencion.sort_values('Porcentaje_Activos', ascending=False)

plt.figure(figsize=(14, 12))

ax = sns.barplot(data=retencion, x='Porcentaje_Activos', y='NOM_CEN', color='lightgreen')

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', padding=3)

plt.title('Porcentaje de Retención (Usuarios Activos vs Registrados) por Centro')
plt.xlabel('Porcentaje de Usuarios Activos (%)')
plt.ylabel('Nombre del Centro')

plt.xlim(0, 110)

plt.tight_layout()
plt.show()
```
