```python
print("Cantidad de niños:", df12["COD_USU"].nunique())
```

```python
unicos  = df12.drop_duplicates(subset=['COD_USU'])
```

```python
sexo_labels = unicos['SEX_USU'].replace({1: 'Hombre', 2: 'Mujer'})

plot_sexo = sns.countplot(x=sexo_labels, hue= sexo_labels)

for container in plot_sexo.containers:
    plot_sexo.bar_label(container)

plt.title('Distribución por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad de Personas')
plt.show()
```

Podemos evidenciar que la cantidad de hombres son 2176 y la canitdad de mujeres son 2190, lo cual evidencia que el servicio de cuidado diurno atendía por igual a niños y niñas.

## Edades de los usuarios 

```python
sns.histplot(data=unicos, x='EDAD_USU', bins=15)

plt.title('Distribución de Edades de los Usuarios')
plt.xlabel('Edad')
plt.ylabel('Cantidad de Personas')

plt.show()
```

```python

edades = unicos[['EDAD_USU']]
```

```python
print("Estadísticas Descriptivas:")

percentiles = edades.quantile([0.25, 0.50, 0.75, 0.90, 0.95, 0.99], numeric_only=True)

print("Media:", edades.mean(numeric_only=True).to_string())
print("Mediana:", edades.median(numeric_only=True).to_string())  
print("Moda:", edades.mode(numeric_only=True).iloc[0].to_string())
print("Varianza:", edades.var(numeric_only=True).to_string())
print("Desviación Estándar:", edades.std(numeric_only=True).to_string())
rango = edades.max(numeric_only=True) - edades.min(numeric_only=True) 
print("Rango:", rango.to_string())

for col in percentiles.columns:
    print(f"\nPercentiles para: {col}")
    print(f"Q1 (25th): {percentiles[col][0.25]:.2f}")
    print(f"Q2 (50th/Median): {percentiles[col][0.50]:.2f}")
    print(f"Q3 (75th): {percentiles[col][0.75]:.2f}")
    print(f"P90: {percentiles[col][0.90]:.2f}")
    print(f"P95: {percentiles[col][0.95]:.2f}")
    print(f"P99: {percentiles[col][0.99]:.2f}")


```
