```python
print("Cantidad de niños:", df12["COD_USU"].nunique())
```

```python
unicos  = df12.drop_duplicates(subset=['COD_USU'])
```

## ACTIVO O INACTIVO

```python
conteo_discapacidad = unicos['EST_ACT'].replace({1: 'Activo', 2: 'Inactivo'}).value_counts().sort_index()

ax = sns.barplot(x=conteo_discapacidad.index, y=conteo_discapacidad.values, color='skyblue')
for container in ax.containers:
    ax.bar_label(container)

plt.title('Cantidad de Activos')
plt.xlabel('¿Activo?')
plt.ylabel('Cantidad de Personas')

plt.show()
```

Podemos ver que más de la mitad de los usarios son activos, por lo que se hará un análisis general y a solo los activos

```python
activos = unicos.query('EST_ACT == 1')
```

```python
sexo_labels = unicos['SEX_USU'].replace({1: 'Hombre', 2: 'Mujer'})
sexo_labels_a = activos['SEX_USU'].replace({1: 'Hombre', 2: 'Mujer'})

fig, axes = plt.subplots(1, 2, figsize=(14, 5))


sns.countplot(x=sexo_labels, hue=sexo_labels, ax=axes[0])

for container in axes[0].containers:
    axes[0].bar_label(container)

axes[0].set_title('Distribución por Sexo de todos los registrados')
axes[0].set_xlabel('Sexo')
axes[0].set_ylabel('Cantidad de Personas')

sns.countplot(x=sexo_labels_a, hue=sexo_labels_a, ax=axes[1])
for container in axes[1].containers:
    axes[1].bar_label(container)

axes[1].set_title('Distribución por Sexo de los activos')
axes[1].set_xlabel('Sexo')
axes[1].set_ylabel('Cantidad de Personas')

plt.tight_layout()
plt.show()
```

Podemos evidenciar que la cantidad de hombres son 2176 y la canitdad de mujeres son 2190, lo cual evidencia que el servicio de cuidado diurno atendía por igual a niños y niñas. Dentro de los usuarios activos, podemos ver que la distribución sigue siendo la misma, pero hay más mujeres que hombres.

## Edades de los usuarios 

```python
edades = unicos[['EDAD_USU']]
edades_activos = activos[['EDAD_USU']]
```

```python
print("Estadísticas Descriptivas de todos los usuarios registrados:")
print()

percentiles = edades.quantile([0.25, 0.50, 0.75, 0.90, 0.95, 0.99], numeric_only=True)

print("Media:", edades.mean(numeric_only=True).to_string())
print("Mediana:", edades.median(numeric_only=True).to_string())  
print("Moda:", edades.mode(numeric_only=True).iloc[0].to_string())
print("Varianza:", edades.var(numeric_only=True).to_string())
print("Desviación Estándar:", edades.std(numeric_only=True).to_string())
rango = edades.max(numeric_only=True) - edades.min(numeric_only=True) 
print("Rango:", rango.to_string())

for col in percentiles.columns:
    print(f"\nPercentiles para: {col}")
    print(f"Q1 (25th): {percentiles[col][0.25]:.2f}")
    print(f"Q2 (50th/Median): {percentiles[col][0.50]:.2f}")
    print(f"Q3 (75th): {percentiles[col][0.75]:.2f}")
    print(f"P90: {percentiles[col][0.90]:.2f}")
    print(f"P95: {percentiles[col][0.95]:.2f}")
    print(f"P99: {percentiles[col][0.99]:.2f}")

print("Estadísticas Descriptivas de los usuarios activos:")
print()

percentiles_a = edades_activos.quantile([0.25, 0.50, 0.75, 0.90, 0.95, 0.99], numeric_only=True)

print("Media:", edades_activos.mean(numeric_only=True).to_string())
print("Mediana:", edades_activos.median(numeric_only=True).to_string())  
print("Moda:", edades_activos.mode(numeric_only=True).iloc[0].to_string())
print("Varianza:", edades_activos.var(numeric_only=True).to_string())
print("Desviación Estándar:", edades_activos.std(numeric_only=True).to_string())
rango = edades_activos.max(numeric_only=True) - edades_activos.min(numeric_only=True) 
print("Rango:", rango.to_string())

for col in percentiles_a.columns:
    print(f"\nPercentiles para: {col}")
    print(f"Q1 (25th): {percentiles_a[col][0.25]:.2f}")
    print(f"Q2 (50th/Median): {percentiles_a[col][0.50]:.2f}")
    print(f"Q3 (75th): {percentiles_a[col][0.75]:.2f}")
    print(f"P90: {percentiles_a[col][0.90]:.2f}")
    print(f"P95: {percentiles_a[col][0.95]:.2f}")
    print(f"P99: {percentiles_a[col][0.99]:.2f}")

```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 15))

# --- Histogramas ---

sns.histplot(data=unicos, x='EDAD_USU', bins=15, ax=axes[0, 0])
axes[0, 0].set_title('Distribución de Edades de los Usuarios Registrados')
axes[0, 0].set_xlabel('Edad')
axes[0, 0].set_ylabel('Cantidad de Personas')

sns.histplot(data=activos, x='EDAD_USU', bins=15, ax=axes[0, 1])
axes[0, 1].set_title('Distribución de Edades de los Usuarios Activos')
axes[0, 1].set_xlabel('Edad')
axes[0, 1].set_ylabel('Cantidad de Personas')


# --- Countplots ---

edades_labels_u = unicos['GRU_ET'].replace({1: '0 - 11 meses', 2: '1 - 5 años', 3: '6 - 11 años', 4: '12 - 17 años', 5: '18 años'})
plot_edad_u = sns.countplot(x=edades_labels_u, hue=edades_labels_u, ax=axes[1, 0])

for container in plot_edad_u.containers:
    plot_edad_u.bar_label(container)

axes[1, 0].set_title('Distribución por Edad de Registrados')
axes[1, 0].set_xlabel('Edad')
axes[1, 0].set_ylabel('Cantidad de Personas')

edades_labels_a = activos['GRU_ET'].replace({1: '0 - 11 meses', 2: '1 - 5 años', 3: '6 - 11 años', 4: '12 - 17 años', 5: '18 años'})
plot_edad_a = sns.countplot(x=edades_labels_a, hue=edades_labels_a, ax=axes[1, 1])

for container in plot_edad_a.containers:
    plot_edad_a.bar_label(container)

axes[1, 1].set_title('Distribución por Edad de Activos')
axes[1, 1].set_xlabel('Edad')
axes[1, 1].set_ylabel('Cantidad de Personas')


# --- Barplots ---

conteo_edades_u = unicos['EDAD_USU'].value_counts().sort_index()
ax_bar_u = sns.barplot(x=conteo_edades_u.index, y=conteo_edades_u.values, color='skyblue', ax=axes[2, 0])

for container in ax_bar_u.containers:
    ax_bar_u.bar_label(container)

axes[2, 0].set_title('Cantidad de Usuarios Registrados por Edad')
axes[2, 0].set_xlabel('Edad')
axes[2, 0].set_ylabel('Cantidad de Personas')

conteo_edades_a = activos['EDAD_USU'].value_counts().sort_index()
ax_bar_a = sns.barplot(x=conteo_edades_a.index, y=conteo_edades_a.values, color='skyblue', ax=axes[2, 1])

for container in ax_bar_a.containers:
    ax_bar_a.bar_label(container)

axes[2, 1].set_title('Cantidad de Usuarios Activos por Edad')
axes[2, 1].set_xlabel('Edad')
axes[2, 1].set_ylabel('Cantidad de Personas')

plt.tight_layout()
plt.show()


```

A raíz de estos gráficos y los resultados de las estadísticas descriptivas, podemos ver que la mayoría de usuarios se encuentran en el rango de 6 a 11 años, pero si hicieramos un rango efectivo serían entre los 3 y 7 años donde se encuentran la mayoría de usuarios, donde la media es de 7 AÑOS y la edad que más se repite es de 4 años. 

Ahora si vemos el gráfico de los usuarios activos, podemos evidenciar que la moda es de 3 años, viendo una distribución parecida a la primera.

## DISCAPACIDAD

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))


conteo_discapacidad = unicos['TIE_DIS'].replace({1: 'Sí, Con CM', 2: 'Si, Sin CM', 3: 'No tiene', 4: 'Pendiente'}).value_counts().sort_index()

ax = sns.barplot(x=conteo_discapacidad.index, y=conteo_discapacidad.values, color='skyblue',ax=axes[0])
for container in ax.containers:
    ax.bar_label(container)

axes[0].set_title('Cantidad de Usuarios Registrados por Discapacidad')
axes[0].set_xlabel('Discapacidad')
axes[0].set_ylabel('Cantidad de Personas')

conteo_discapacidad_a = activos['TIE_DIS'].replace({1: 'Sí, Con CM', 2: 'Si, Sin CM', 3: 'No tiene', 4: 'Pendiente'}).value_counts().sort_index()

ax_a = sns.barplot(x=conteo_discapacidad_a.index, y=conteo_discapacidad_a.values, color='skyblue',ax=axes[1])
for container in ax_a.containers:
    ax_a.bar_label(container)

axes[1].set_title('Cantidad de Usuarios Activos por Discapacidad')
axes[1].set_xlabel('Discapacidad')
axes[1].set_ylabel('Cantidad de Personas')

plt.tight_layout()
plt.show()
```

Se puede evidenciar que la presencia de discapacidades es casi nula en los registrados y en los activos.
